```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Exception Handling
try:
    # Read the CSV file
    df = pd.read_csv('olympic_medals.csv')
except FileNotFoundError:
    print('File not found.')
    exit()
except pd.errors.EmptyDataError:
    print('File is empty.')
    exit()
except ValueError:
    print('Value error occurred, check the values in your data.')
    exit()
except KeyError:
    print('Key error occurred, check the column names in your data.')
    exit()
except Exception as e:
    print(f'An unexpected error occurred: {e}')
    exit()

# File Handling
summary = df.describe()
file_path = 'summary_olympic_analysis.txt'   #Define the file path
try:
    with open(file_path, 'w') as f:
        f.write(summary.to_string(header=True, index=True))
    print(f"Summary statistics saved to '{file_path}' successfully.")
except IOError as e:
    print(f"Error: Unable to write summary statistics to '{file_path}'. {e}")


# Display initial dataset information
print("Original dataset")
print(df)
print(df.shape)

# Handle missing values
df.dropna(inplace=True)
print("After Handling Missing Values")
print(df)

# Handle duplicate values
df.drop_duplicates(inplace=True)
print("After Handling Duplicate Values")
print(df)
print(df.shape)

# Plotting
x_column = 'Discipline'  # Replace with actual column name you want to use
```

```python
# BARH CHART
plt.figure(figsize=[20,20])
plt.barh(
    df[x_column].head(60).value_counts().index, 
    df[x_column].head(60).value_counts().values, 
    color=('red', 'black'),
    label='Winners'
)
plt.show()
```

```python
# PIE CHART
plt.figure(figsize=[5,5])
plt.pie(df[x_column].value_counts().values, autopct='%1.1f%%', labels=df[x_column].value_counts().index)
plt.show()
```

```python
# BAR CHART
plt.figure(figsize=[5,5])
plt.bar(
    df[x_column].head(60).value_counts().index, 
    df[x_column].head(60).value_counts().values, 
    color=('purple', 'black'),
)
plt.ylabel(x_column)
plt.show()
```

```python
# SCATTER CHART
plt.figure(figsize=[10,5])
plt.scatter(df.index[:60], df[x_column].head(60))
plt.xlabel('Index')
plt.ylabel(x_column)
plt.title('Scatter Plot')
plt.show()
```

```python
# LINE CHART
plt.figure(figsize=[10,5])
plt.plot(df.index[:60], df[x_column].head(60), marker='o', linestyle='-', color='r', label='Line 1')
plt.title('Line Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# AREA CHART
plt.figure(figsize=[10,5])
plt.fill_between(df.index[:60], df[x_column].head(60), color='purple', alpha=0.4, label='Area 1')
plt.plot(df.index[:60], df[x_column].head(60), color='blue', alpha=0.6, linewidth=2)
plt.title('Area Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
